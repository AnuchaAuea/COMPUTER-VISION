```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '../.././images/thai_text1.jpg'

# Load the image

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# display the image using matplotlib
plt.imshow(image)

```

# Homework ->    

```python
mask = np.zeros_like(image, dtype=np.uint8)
plt.imshow(mask)

cv2.rectangle(mask, (130, 550), (550, 450), (255, 255, 255), -1)
plt.imshow(mask)

masked_image = cv2.bitwise_and(image, image, mask=mask)
plt.imshow(masked_image)

```

```python
ret, result = cv2.threshold(masked_image, 200, 255, cv2.THRESH_BINARY)

# 
plt.imshow(result, cmap='gray')
```
